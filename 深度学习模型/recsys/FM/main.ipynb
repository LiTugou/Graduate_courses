```python
import pandas as pd
import numpy as np
```

```python
datapath1="./ml-latest-small/ratings.csv"
data1=pd.read_csv(datapath1)
datapath2="./ml-latest-small/movies.csv"
data2=pd.read_csv(datapath2)
```

```python
print(data1.info())
print(data2.info())
```

## 对类别One-Hot
值 "Action|Crime|Horror|Mystery|Thriller"

```python
genres=data2["genres"]
cat_dict={}
cat_df=data2[["movieId"]]
for i in range(data2.shape[0]):
    catogories=genres.iloc[i].lower().split("|")
    for cat in catogories:
        if cat not in cat_dict:
            cat_dict.setdefault(cat,len(cat_dict))
            new_cat=pd.Series(np.zeros((data2.shape[0],),dtype=np.int8),name=cat)
            cat_df=pd.concat([cat_df,new_cat],axis=1)
    cat_df.loc[i,catogories]=1
```

```python
cat_df.sum(axis=0)
```

## user和movies one-hot
不理解，为什么要one-hot

```python
one_hot_user = pd.get_dummies(data1["userId"].values,prefix="user")
one_hot_mov = pd.get_dummies(data1["movieId"].values,prefix="mov")
```

## 合并类别one-hot数据

```python
data=data1.join(one_hot_user).join(one_hot_mov)
data=pd.merge(data,cat_df,on="movieId",how="left")
data=data.drop(columns=["userId","movieId","timestamp"])
data
```
